```python
import numpy as np
import pandas as pd
import logging
```

```python
# Set up logging
logging.basicConfig(level=logging.INFO, format='%(asctime)s - %(levelname)s - %(message)s')

data = {
    'index': np.arange(1, 201),
    'amount': np.random.randint(1, 100, size=200)  
}
df = pd.DataFrame(data)

target_sum = 23

```

```python
def find_combinations(df, target_sum):
    result = []

    def backtrack(start, current_sum, current_indices):
        try:
            if current_sum == target_sum:
                result.append(current_indices[:])
                logging.info(f"Found combination: {current_indices}")
                return
            if current_sum > target_sum:
                return
            
            for i in range(start, len(df)):
                current_indices.append(df['index'][i])
                backtrack(i + 1, current_sum + df['amount'][i], current_indices)
                current_indices.pop()
        except Exception as e:
            logging.error(f"Error occurred: {e}")
    
    backtrack(0, 0, [])
    return result
```

```python
try:
    combinations = find_combinations(df, target_sum)
    df_new = []
    if combinations:
        for combination in combinations:
            print(f"Indices that sum to {target_sum}: {combination}")
            matching_rows = df[df['index'].isin(combination)]
            df_new.append(matching_rows)
            print(matching_rows)
    else:
        raise ValueError("No combinations match the target sum.")
except ValueError as ve:
    logging.info(f"No combinations found that sum to {target_sum}.")
```
